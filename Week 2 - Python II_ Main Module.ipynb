### For loops

The understanding of loops is vital to programming. <span style="background-color: #AFEEEE">**For Loops**</span> offer a way to perform repetitive tasks with minimal code. It is a control structure that's used to execute a block of code repeatedly. Imagine having to print numbers from 1 to 1000 without a loop. That is a lot of typing (or copy and pasting). With loops, such tasks become trivial.

But where would such repetition be common? One of the most common applications is for iterating through a list and applying certain operations to each of the entries. By the end of this section of the module, we would like to be able to calculate the average of a column/variable within our heart failure dataset itself. This means that we would like to take the sum of the values in a column, platelet levels, for example, and then divide this sum by the number of values in the column itself. By summing the entries, this requires us to iterate through the column; hence, we can leverage the characteristic of a for loop to help us achieve this calculation.

But first, let's touch on some of the basics.

#### **Simple For Loop Structure**

````python
colors = ['red', 'blue', 'yellow', 'green', 'purple']
for color in colors:
    print(color)
````

<img src="image/image-2.gif" alt="loop structure demo" style="width: 600px;"/>

A great aspect of Python code is that its syntax is generally easy to read, but let's walk through this snippet of code. In plain English, it essentially means "We have a list of colors that includes `'red'`, `'blue'`, `'yellow'`, `'green'`, and `'purple'`. For each `color` in the list, display the name of that color."


In this example:
- `colors` is a list — a type of iterable object — which means it contains multiple items that can be looped through
- `color` is a local variable used within the loop. It represents the current item from the list during each iteration — it’s just a placeholder used while the loop runs.
So, the loop goes through each item in the `colors` list one by one, and prints it out.

In general, a loop structure looks like `for x in y`, where `x` is a temporary variable and `y` can be a sequence, an iterable object, or a number range. Although you can name `x` anything you like, it's recommended to name it according to the meaning it stands for in the for loop. For example, 'index' if it stands for an index, 'color' if it refers to a color, or 'char' if it refers to a character.

#### **Looping Using Indices**

While directly looping through items in a list is readable, another more common approach is to loop through the indices of the list. This means accessing each item by its position (or index) in the list.

````python
for i in range(len(colors)):
    print(colors[i])
````

We can also use the index as a means of producing calculations. Imagine you want the square of each number from 0 to 9. We can complete that in a for loop where we multiply the variable `i` by itself and print the result each iteration.

````python
for i in range(10):
    print(i*i)
````

#### **Calculating Averages Using For Loops**

What is a practical application of loops in Python? Consider a column of data. How might you determine the sum of all its values? This column can be visualized as a list of values in Python. By iterating through each value, we can accumulate them to get the total sum. Let's refer back to our heart failure dataset and calculate the sum of all platelet levels. Here's an example demonstrating this approach:

````python
# Extract 'platelets' column as a list
platelets_list = df['platelets'].tolist()
# First 5 entries
platelets_list[:5]
````

It seems that from our initial dataframe, we took the `platelets` column and stored its values in the list called `platelets_list`. Now, let's use a for loop to calculate the sum of all the values. We will store this in a variable called `platelets_sum`.

````python
# Initialize sum to 0
platelets_sum = 0

# Use a for loop to sum all the values
for value in platelets_list:
    platelets_sum += value  # This is equivalent to platelets_sum = platelets_sum + value
````

First, a variable named `platelets_sum` is created and set to 0. This variable will be used to accumulate or store the sum of the values as we iterate through the list. A for loop is used to iterate over each value in the `platelets_list`. Each value is added to `platelets_sum` using the `+=` operator. By the end of the loop, `platelets_sum` contains the total sum of all values in the list. What can we do with this value, and why did we calculate it? Well, we can divide it by the number of entries in the list to get the average platelet level in the patients.

````python
platelets_avg = platelets_sum/len(platelets_list)

print(platelets_avg)
````

Now, we have our average platelet levels for all our observations, which is 262458.91 kiloplatelets/mL. With this knowledge, we can easily adapt the procedure to calculate the average of a different column. Let's assume we want to know the average percentage of blood leaving the heart as described in our dataset. We would be looking at the ejection fraction.

### Functions

<span style="background-color: #AFEEEE">**Functions**</span> enable us to bundle code into reusable packages, allowing for modularity and reduced redundancy in programming. In Python, a function is a block of code that only runs when it is called. You can pass data into a function, and it can return data as a result. In fact, you've used functions before. The code snippet `print()` is a function. Of course, there is underlying code that gets called, which displays the prompt. We do not personally have to take time to write out this code.

In the previous for loop section, we have provided an example of how to calculate the average of a column in the dataset, specifically the average platelet level. After such an example, you were tasked with writing a similar function again, but to perform the same procedure on the ejection fraction column in the dataset. You may have observed how a majority of the code structure is the same and feels redundant and trivial to implement. By the end of this section on functions, you will know how to reduce this redundancy and avoid code duplication; hence, saving us time in our implementation and improving the readability of our code.

So, how are functions defined?

````python
def greet():
    return "Hello, World!"

hello = greet()
print(hello)
````

To define a function, we use the `def` keyword. `greet` here is the name of the function. The empty brackets are where parameters would go, which, for our function, there are none. We will return the string "Hello World!" at the end of our function. In the example above, we then stored this returned value in the variable `hello`.

To call a function like you have been doing with `print()`, you specify the name of the function you are calling, and place any parameters necessary in the brackets. What would calling `greet()` do?

````python
greet()
````

When we run the code cell, we can see that the string is returned, indicated by the single quotation marks. This is different from the first time we used the `print()` function, where we specifically passed the string we wanted to print into it.

If we want to allow for some customization as to what `greet()` would display every time we call it, we give it a parameter.

<span style="background-color: #AFEEEE">**Parameter**</span>: Also known as an argument, is a named variable used in the declaration of a function. It acts as a placeholder for actual values that will be provided or passed in when the function is called.

In the example below, `first_name` is a parameter, and in this case, we would like the name to be a string. Essentially,  parameters allow functions to be flexible and general by enabling them to operate on different data depending on the arguments provided during function invocation.

````python
def new_greet(first_name, last_name):
    return "Hello " + first_name + " " + last_name + "!"
````

````python
new_greet('John', 'Doe')
````

In essence, function parameters allow for abstraction, enabling a single function to operate on a variety of data, perform different tasks, or produce different outcomes based on the arguments passed during its invocation.

### Introduction to CSV files

<span style="background-color: #AFEEEE">**Comma Separated Values (CSV)**</span> is a widely accepted data format, one which we are using for our example dataset. CSV files are often used for data storage, data transfer, and data analysis. Reading and writing CSV files in a way that correctly handles field delimiters, line endings, special characters, and varied formatting can be tricky. At its core, a CSV file is just a text file that uses specific structuring to represent tabular data. Think of it like a basic spreadsheet, but instead of being in a proprietary format like Microsoft's .xlsx or Google Sheets' format, it's in a plain-text format that can be read by many software programs. We have already encountered this sort of data. Previously, our examples were derived from the CSV file with the assistance of the `pandas` module.

A CSV file's structure is straightforward:

- Each line in the file represents a row of the table.
- Within each line, values (or cells in our table analogy) are separated by a delimiter, most commonly a comma (hence the name).
- The first row often represents headers or column names, but this isn't a strict rule.

Example:

```
Name, Age, Occupation
Alice, 28, Engineer
Bob, 22, Data Analyst
Charlie, 34, Designer
```

This represents a table with three rows and three columns.

Of course, before conducting any analysis, a quick visual quality check of the dataset is always warranted. Microsoft Excel is a popular tool for viewing, editing, and creating CSV files. Here's how you can use Excel to open a CSV; we will be using our **'hf_data_raw.csv'** file to demonstrate:

- Open Microsoft Excel.
- Go to File > Open and locate the 'hf_data_raw.csv' file in this week's (Week 2 Programming Concepts) folder (look in the data/ subfolder).
- Once selected, the file will open in Excel, and you can see the tabular structure.

Once our CSV file is opened in Excel, you can treat it like any other spreadsheet. You may edit cells, add rows/columns, apply formulas, etc. However, remember that a CSV file is a much simpler file type than xlsx, so sometimes Excel may have trouble formatting its data, and changes made through Excel may not be properly saved in the CSV format. After editing, you can save it as a CSV again by selecting File > Save As and choosing the CSV format. Just keep the name as is if you would like to overwrite the previous version.

When you open a CSV in Excel and see all the data in the first column, it might be because the CSV uses a different delimiter (like a semicolon or a tab). You can use Excel's "Text to Columns" feature in the Data tab to correctly separate the values. Usually, there will be a window that asks you about the formatting of the file when you first open it. Make sure you check to see that Excel has the correct values in each cell. Lucky for us, this is not an issue with 'hf_data_raw.csv'.

## Conclusion

In conclusion, this week's module imparted knowledge on dictionaries, for loops, functions, and the crucial aspect of working with CSV files. By assimilating these concepts, we establish a foundational understanding of data manipulation and processing in Python. To summarize:

<span style="background-color: #AFEEEE">**Dictionaries**</span>: These are unordered collections in Python used to store data in key-value pairs. They offer rapid data access based on unique keys. This feature distinguishes them from lists, which are ordered and utilize index-based access. Dictionaries employ curly braces `{}` for creation. Data is accessed via their corresponding keys, and one can easily add or modify key-value pairs using assignment.

<span style="background-color: #AFEEEE">**For Loop**</span>: A control structure in Python, the for loop iterates over sequences (like lists, tuples, or strings) or other iterable objects. This enables executing a specific code block multiple times, once per sequence item. The `range()` function, often used with for loops, generates a number sequence and helps keep track of iteration counts.

<span style="background-color: #AFEEEE">**Functions**</span>: These are segments of reusable code tailored to execute specific tasks. They can accept parameters and return results. Defined using the `def` keyword, functions are invoked by their name, accompanied by parentheses. Parameters act as placeholders for values that a function accepts. The real values supplied to these placeholders are termed arguments. To send back results to the function caller, the `return` keyword is used.

<span style="background-color: #AFEEEE">**CSV File Format**</span>: The CSV (Comma-Separated Values) file format is crucial in data handling tasks. It's a simple, plain-text format that represents tabular data, making it universally readable by various software and programming languages. Within a CSV file, data is structured in rows, often with the first row serving as headers. These headers act as identifiers for the data columns, facilitating organized data processing and manipulation. Understanding the structure and characteristics of the CSV format is foundational to effective data management.

By mastering these concepts, one steps closer to becoming proficient in data-driven tasks using Python, paving the way for more advanced topics in the future.
